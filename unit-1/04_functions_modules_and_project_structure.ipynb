## Part 1: Functions
Functions keep code DRY (Don't Repeat Yourself) and structured. Let's cover parameters, return statements, default arguments, and `*args`/`**kwargs`.


```python
def format_prompt(instruction, user_input, model='default-llm'):
    """
    Generates a structured system-style instruction prompt.
    """
    return f"""System instruction: {instruction}\nUser Input: {user_input}\nTarget Model: {model}"""

# Testing function calls
prompt_1 = format_prompt("Summarize in 3 bullet points", "Python is popular because of libraries.")
prompt_2 = format_prompt("Translate to French", "Hello, how are you?", model='translator-v2')

print('Prompt 1:\n', prompt_1)
print('\nPrompt 2:\n', prompt_2)
```

## Part 2: Understanding *args and **kwargs
- `*args` collects positional arguments as a tuple.
- `**kwargs` collects keyword arguments as a dictionary.
This is highly useful when forwarding arguments to downstream API configuration functions.


```python
def make_api_call(prompt, **kwargs):
    print(f'Sending prompt: {prompt}')
    print('Configuration parameters passed:')
    for key, value in kwargs.items():
        print(f' - {key}: {value}')

make_api_call('What is AI?', temperature=0.7, max_tokens=100, top_p=0.9)
```

## Hands-On Exercise
**Task**: Write a function `clean_input(text, lowercase=True, remove_punctuation=False)` with a docstring.
It should remove leading/trailing whitespace. If `lowercase` is True, it converts text to lowercase. If `remove_punctuation` is True, it strips `.`, `,`, `!`, and `?` characters.
Verify the function output with several test sentences.


```python
def clean_input(text, lowercase=True, remove_punctuation=False):
    """
    Cleans input text by stripping whitespace, conditionally converting to lowercase,
    and conditionally removing trailing punctuation characters.
    """
    # Clean whitespaces
    cleaned = text.strip()
    
    if lowercase:
        cleaned = cleaned.lower()
        
    if remove_punctuation:
        for p in ['.', ',', '!', '?']:
            cleaned = cleaned.replace(p, '')
            
    return cleaned

# Test inputs
dirty_text = '  Attention! This text HAS some whitespaces and punctuation.? '
print('Original:', dirty_text)
print('Default Cleaned:', clean_input(dirty_text))
print('Fully Processed:', clean_input(dirty_text, lowercase=True, remove_punctuation=True))
```
