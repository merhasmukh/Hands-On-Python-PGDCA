## Part 1: What is a Virtual Environment?
A virtual environment is a self-contained directory tree that contains a Python installation for a particular version of Python, plus a number of additional packages.
This keeps dependencies required by different projects separate. For instance, Project A might need `requests` v2.25, while Project B needs `requests` v2.31.

### Commands to create and activate a virtual environment (in terminal):
1. **Create**: `python -m venv ai_env`
2. **Activate**:
   - **Mac/Linux**: `source ai_env/bin/activate`
   - **Windows (cmd)**: `ai_env\Scripts\activate.bat`
   - **Windows (PowerShell)**: `ai_env\Scripts\Activate.ps1`
3. **Deactivate**: `deactivate`


## Part 2: Working with pip (Package Installer for Python)
Once activated, let's explore pip commands inside the environment:
- `pip install <package>`: Install package
- `pip install --upgrade <package>`: Upgrade package
- `pip uninstall <package>`: Remove package
- `pip list`: List installed packages
- `pip freeze > requirements.txt`: Output installed packages to a dependency file
- `pip install -r requirements.txt`: Install all dependencies listed in the file


````python
# Let's check which pip and python are currently active in the notebook kernel
import sys
import subprocess
print('Active Python path:', sys.executable)
# Run a shell command to list packages installed in current kernel
result = subprocess.run([sys.executable, '-m', 'pip', 'list'], capture_output=True, text=True)
print(result.stdout[:500]) # Display first 500 chars
````

## Part 3: Folder Structure Best Practices for AI Projects
A standard enterprise AI project follows this template:
```text
my_ai_project/
├── ai_env/              # Ignored in git (.gitignore)
├── data/                # Data inputs/outputs (ignored if large)
├── src/                 # Source code modules
│   ├── __init__.py
│   ├── main.py
│   └── utils.py
├── .env                 # API keys & secret config (DO NOT COMMIT!)
├── .env.example         # Template env file with placeholders
├── requirements.txt     # List of library dependencies
└── README.md            # Documentation
```


## Hands-On Exercise
**Task**: Let's write code to programmatically create the project directory skeleton described above using Python's standard `os` library.
This will prepare you for folder setup best practices.


````python
import os

# TODO: Define directories to create inside a folder named 'my_ai_project'
project_root = 'my_ai_project'
dirs_to_create = [
    os.path.join(project_root, 'data'),
    os.path.join(project_root, 'src'),
]

for d in dirs_to_create:
    os.makedirs(d, exist_ok=True)
    print(f'Created directory: {d}')

# Create empty source files and README.md
files_to_create = [
    os.path.join(project_root, 'src', 'main.py'),
    os.path.join(project_root, 'src', 'utils.py'),
    os.path.join(project_root, 'README.md'),
    os.path.join(project_root, '.env.example'),
    os.path.join(project_root, 'requirements.txt'),
]

for f in files_to_create:
    with open(f, 'w') as file:
        if 'README' in f:
            file.write('# My AI Project\nThis is a standard template project.')
    print(f'Created file: {f}')
````
