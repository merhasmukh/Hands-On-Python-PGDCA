## Part 3: Troubleshooting Common First Errors
As beginners, you will encounter error messages like `SyntaxError` or `NameError`. Don't panic! Let's understand them.

### 1. NameError (Variables not defined)
Try running the cell below. It will crash because `api_key` has not been defined yet.


```python
try:
    print(api_key)
except NameError as e:
    print('Caught expected NameError:', e)
```

### 2. SyntaxError
SyntaxError occurs when Python parser cannot compile code due to grammatical mistakes (e.g., mismatched quotes, unclosed brackets).


```python
try:
    # Code with invalid syntax:
    # print('Hello)
    eval("print('Hello") # triggering syntax error programmatically
except SyntaxError as e:
    print('Caught expected SyntaxError:', e)
```
