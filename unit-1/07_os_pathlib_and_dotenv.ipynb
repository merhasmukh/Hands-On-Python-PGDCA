## Part 1: Path Handling with os vs pathlib
Python's standard library offers two ways to handle filesystem paths:
- `os.path`: legacy string-based manipulations.
- `pathlib.Path`: modern, object-oriented path library.


```python
import os
from pathlib import Path

# Check current working directory
current_dir_os = os.getcwd()
current_dir_pathlib = Path.cwd()
print('OS current dir:', current_dir_os)
print('Pathlib current dir:', current_dir_pathlib)

# Joining path components safely across OS (Windows uses \, Mac/Linux uses /)
data_folder_os = os.path.join(current_dir_os, 'data')
data_folder_pathlib = current_dir_pathlib / 'data'

print('OS data path:', data_folder_os)
print('Pathlib data path:', data_folder_pathlib)
print('Does path exist?', data_folder_pathlib.exists())
```

## Part 2: Listing Files in a Directory
Let's list files in the `data/` directory and filter them by extensions (e.g. only `.json`).


```python
data_dir = Path('data')
print('All files in data/ directory:')
for item in data_dir.iterdir():
    if item.is_file():
        print(f' - {item.name} ({item.suffix})')

print('\nOnly JSON files in data/ directory:')
json_files = list(data_dir.glob('*.json'))
for j in json_files:
    print(f' - {j.name}')
```

## Part 3: Managing Secrets with python-dotenv
It is a critical security violation to hardcode API Keys or authentication credentials in your codebase.
Instead, we write variables to a `.env` file that is ignored by Git, and load them using `python-dotenv`.

Let's write a mock `.env` file for demonstration purposes.


```python
%%writefile .env
OPENAI_API_KEY=sk-proj-mock1234567890abcdef
LLM_MODEL=gpt-4o-mini
DEBUG_MODE=True
```

Now let's load this environment configuration using `dotenv.load_dotenv()`.


```python
from dotenv import load_dotenv

# Load variables from .env file into environment variables
load_dotenv()

# Retrieve using standard os.getenv
api_key = os.getenv('OPENAI_API_KEY')
model = os.getenv('LLM_MODEL')
debug_mode = os.getenv('DEBUG_MODE')

print('Loaded API Key:', api_key[:12] + '...' if api_key else None)
print('Loaded Model:', model)
print('Debug Mode:', debug_mode)
```
