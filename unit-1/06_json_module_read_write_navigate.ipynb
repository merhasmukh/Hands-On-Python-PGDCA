# Practical P6: The json Module: Read, Write, Navigate
**Learning Outcome**: Read and write JSON files and objects fluently — the native format of all internet APIs.

## Part 1: Parsing and serializing JSON in memory
- `json.loads(string)`: Parses JSON-formatted string to a Python dictionary.
- `json.dumps(dict)`: Converts (serializes) Python dictionary to JSON-formatted string.


```python
import json

# A JSON string received from a web server
json_string = '{"model": "gpt-4", "temperature": 0.5, "max_tokens": 100}'
print('Type of raw string:', type(json_string))

# Parse string to Python dict
config_dict = json.loads(json_string)
print('Type after load:', type(config_dict))
print('Accessing model property:', config_dict['model'])

# Serialize Python dict back to JSON string with pretty indenting
new_json_str = json.dumps(config_dict, indent=4)
print('Pretty-printed JSON String:\n', new_json_str)
```

## Part 2: Reading and Writing JSON files
- `json.load(file_object)`: Decodes JSON file directly to Python dict.
- `json.dump(dict, file_object)`: Writes dict directly to a JSON file.


```python
# Let's load the sample API response JSON file we created in our data folder
file_path = 'data/sample_api_response.json'

with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

print(f'Loaded content from {file_path}:')
print('Model used:', data['model'])
print('Assistant response content:', data['choices'][0]['message']['content'])
```

## Hands-On Exercise
**Task**: Load the `data/sample_api_response.json` file. Extract the following 5 fields:
1. `id`
2. `model`
3. `choices[0].message.content` (the text response)
4. `usage.total_tokens`
5. `metadata.latency_ms`

Save these extracted attributes as a new summary dictionary, and write it to `data/response_summary.json` with an indent of 2.


```python
# TODO: Complete the extraction and file write
with open('data/sample_api_response.json', 'r') as f:
    full_resp = json.load(f)

summary = {
    'id': full_resp['id'],
    'model': full_resp['model'],
    'response_content': full_resp['choices'][0]['message']['content'],
    'total_tokens': full_resp['usage']['total_tokens'],
    'latency_ms': full_resp['metadata']['latency_ms']
}

out_path = 'data/response_summary.json'
with open(out_path, 'w') as f:
    json.dump(summary, f, indent=2)

print(f'Successfully wrote summary to {out_path}:')
with open(out_path, 'r') as f:
    print(f.read())
```
