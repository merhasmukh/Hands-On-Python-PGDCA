# Practical P9: pandas Part 2: Cleaning & Preprocessing Text
**Learning Outcome**: Preprocess raw, messy text data to make it clean and ready for an AI model pipeline.

## Part 1: Handing Missing Values
Null text entries cause code errors inside embeddings API scripts. We must drop them or replace them with defaults.
- `df.dropna(subset=['col'])`: Drops rows with null values in specified column.
- `df.fillna({'col': 'default'})`: Replaces null values with fallback strings.


```python
import pandas as pd
df = pd.read_csv('data/customer_reviews.csv')
print('Initial count of null values in each column:')
print(df.isnull().sum())

# Clean missing review_text with a default placeholder or drop
df['review_text'] = df['review_text'].fillna('')
print('Null values in review_text after replacement:', df['review_text'].isnull().sum())
```

## Part 2: Vectorized String Operations
Pandas provides vectorized operations for text processing under the `str` accessor:
- `df['col'].str.strip()`: Remove leading/trailing whitespaces.
- `df['col'].str.lower()`: Lowercase characters.
- `df['col'].str.replace(old, new)`: Substitute patterns.
- `df['col'].str.len()`: Count character lengths.


```python
# Let's clean the review text
df['clean_review'] = df['review_text'].str.strip().str.lower()
print('Original vs Cleaned Review Text:')
print(df[['review_text', 'clean_review']].head(4))
```

## Part 3: Feature Extraction (e.g. Word Counts)
We can compute word count metrics for analytics or filters prior to LLM pricing checks.


```python
# Compute word count for each clean review
df['word_count'] = df['clean_review'].apply(lambda x: len(x.split()) if x else 0)
print(df[['clean_review', 'word_count']].head(5))
```

## Hands-On Exercise
**Task**: Clean a messy customer review CSV by implementing the complete preprocessing flow:
1. Drop rows where `customer_name` is missing.
2. Strip all whitespaces and convert `review_text` to lowercase.
3. Create a column `char_length` holding the character length of the cleaned reviews.
4. Save the cleaned DataFrame to `data/cleaned_reviews.csv` without indexing headers.


```python
# TODO: Complete preprocessing recipe
df_raw = pd.read_csv('data/customer_reviews.csv')

# Drop rows missing customer_name
df_cleaned = df_raw.dropna(subset=['customer_name']).copy()

# Handle review_text missing/whitespaces
df_cleaned['review_text'] = df_cleaned['review_text'].fillna('')
df_cleaned['review_text'] = df_cleaned['review_text'].str.strip().str.lower()

# Add char length
df_cleaned['char_length'] = df_cleaned['review_text'].str.len()

# Save
out_path = 'data/cleaned_reviews.csv'
df_cleaned.to_csv(out_path, index=False)

print('Preprocessed reviews summary:')
print(pd.read_csv(out_path).head(3))
```
