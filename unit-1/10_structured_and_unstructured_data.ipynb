## Part 1: Unstructured Data (Raw Text File)
Unstructured data like PDF files, word files, and raw text does not have a predefined format. In modern AI systems, we split long documents into smaller parts ('chunks') so they fit inside an LLM's prompt context length window limit.


```python
from pathlib import Path

# Read our sample article file
article_path = Path('data/sample_article.txt')
with open(article_path, 'r', encoding='utf-8') as f:
    text_content = f.read()

print(f'Total length of raw article: {len(text_content)} characters')
print('First 200 characters:\n', text_content[:200])
```

## Part 2: text Chunking (Word Count Threshold)
Let's write a python function to chunk text into lists of strings, where each chunk has a maximum of `N` words.


```python
def chunk_text_by_words(text, chunk_size=50):
    words = text.split()
    chunks = []
    for i in range(0, len(words), chunk_size):
        chunk = ' '.join(words[i:i + chunk_size])
        chunks.append(chunk)
    return chunks

chunks = chunk_text_by_words(text_content, chunk_size=70)
print(f'Created {len(chunks)} chunks.')
print('Chunk 1 (70 words):\n', chunks[0])
print('-'*40)
print('Chunk 2 (70 words):\n', chunks[1])
```

## Hands-On Exercise
**Task**: Create a chunking workflow:
1. Read the article `data/sample_article.txt`.
2. Chunk the text content into sections of exactly 100 words each.
3. Save each chunk to a separate JSON file named `data/chunk_1.json`, `data/chunk_2.json`, etc. containing matching dictionary format:
   `{'chunk_id': index, 'word_count': length, 'content': text}`.


```python
# TODO: Complete chunking recipe
with open('data/sample_article.txt', 'r') as f:
    article = f.read()

words = article.split()
chunk_size = 100
chunk_idx = 1

for i in range(0, len(words), chunk_size):
    chunk_words = words[i:i + chunk_size]
    chunk_text = ' '.join(chunk_words)
    
    chunk_data = {
        'chunk_id': chunk_idx,
        'word_count': len(chunk_words),
        'content': chunk_text
    }
    
    chunk_filename = f'data/chunk_{chunk_idx}.json'
    with open(chunk_filename, 'w') as out_f:
        json.dump(chunk_data, out_f, indent=2)
    
    print(f'Saved chunk {chunk_idx} to {chunk_filename}')
    chunk_idx += 1
```
