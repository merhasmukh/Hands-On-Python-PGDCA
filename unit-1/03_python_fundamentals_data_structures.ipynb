## Part 1: Lists
Lists are ordered, mutable sequences. We use lists to store sequential tokens, messages in a chat history, or dataset records.


```python
# Create a chat transcript log
chat_history = ['Hello assistant!', 'Hi user! How can I help?', 'What is 2+2?', '2+2 is 4.']
print('Original:', chat_history)

# Slicing lists
print('First 2 messages:', chat_history[:2])
print('Last message:', chat_history[-1])

# Appending data
chat_history.append('Thank you!')
print('Updated list:', chat_history)
```

## Part 2: Dictionaries (JSON-like)
Dictionaries store data in key-value pairs. Large Language Model (LLM) APIs (like OpenAI, Gemini, Claude) return JSON structures which Python parses as nested dictionaries.


```python
# Mock LLM settings dictionary
model_config = {
    'name': 'gemini-1.5-pro',
    'temperature': 0.2,
    'safety_settings': {
        'harassment': 'block_medium_and_above',
        'hate_speech': 'block_low_and_above'
    }
}
print('Model configuration name:', model_config['name'])
print('Hate speech safety setting:', model_config['safety_settings']['hate_speech'])
```

## Part 3: List Comprehensions
List comprehensions provide a concise way to create lists. They are widely used for filtering or transforming data before passing it into models.


```python
sentences = ['   data preparation is key ', ' clean this text  ', 'no-whitespace-issues']
# Let's clean the sentences by stripping whitespace
cleaned_sentences = [s.strip() for s in sentences]
print('Cleaned:', cleaned_sentences)
```
