## Part 1: Introduction to pandas
Pandas is the default standard library used for data cleaning and manipulation in Python.
- `Series`: 1D labeled array.
- `DataFrame`: 2D tabular structure with labeled rows and columns (like an Excel worksheet or SQL table).

Let's import pandas and load a customer review CSV dataset.


```python
import pandas as pd

# Load the generated customer reviews CSV dataset
df = pd.read_csv('data/customer_reviews.csv')

# Check shape (number of rows, number of columns)
print('Dataset Shape:', df.shape)
```

## Part 2: Quick Inspection Methods
- `df.head(n)`: Views first n rows.
- `df.tail(n)`: Views last n rows.
- `df.info()`: Summary metadata of data types, missing values.
- `df.describe()`: Statistical description of numeric columns.


```python
# View first 5 rows
print('--- HEAD ---')
print(df.head(5))

# View dataset summary
print('\n--- INFO ---')
df.info()
```

## Part 3: Selecting and Filtering Data
- Selecting a column: `df['column_name']` or `df[['col1', 'col2']]`
- Filtering rows based on condition: `df[df['rating'] > 3]`


```python
# Value count distribution for sentiment labels
print('Sentiment distribution:')
print(df['sentiment'].value_counts())

# Convert rating to numeric (handling errors gracefully as coercion)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Filter reviews with rating >= 4
high_ratings = df[df['rating'] >= 4]
print(f'Number of reviews rated 4 or 5: {len(high_ratings)}')
print(high_ratings[['customer_name', 'rating', 'sentiment']].head(3))
```

```python
# TODO: Complete exploration task
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
avg_rating = df['rating'].mean()
print(f'Average customer rating: {avg_rating:.2f}')

# Find longest review text
# Fill missing review texts with empty string
df['review_text'] = df['review_text'].fillna('')
longest_idx = df['review_text'].str.len().idxmax()
longest_review = df.loc[longest_idx]

print('\nLongest Review Info:')
print('Customer Name:', longest_review['customer_name'])
print('Rating:', longest_review['rating'])
print('Length:', len(longest_review['review_text']))
print('Content:', longest_review['review_text'])
```
