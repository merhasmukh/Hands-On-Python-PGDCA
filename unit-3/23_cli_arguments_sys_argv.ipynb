# Practical P23: Parsing Command Line Arguments with `sys.argv`
**Learning Outcome**: Understand how values are passed to command line scripts and parse basic flags using Python's standard `sys` module.

## Part 1: What is `sys.argv`?
`sys.argv` is a list in Python that contains the command-line arguments passed to the script.
- `sys.argv[0]` is always the script name.
- `sys.argv[1:]` contains the actual arguments passed after the script name.


```python
import sys
print('Script path:', sys.argv[0])
print('List of arguments:', sys.argv)
```

## Hands-On Exercise
**Task**: Write a Python script `echo_cli.py` using `sys.argv` that prints the number of arguments passed, and outputs each argument on a new line. Verify it by executing it in the terminal.


```python
%%writefile echo_cli.py
import sys

args = sys.argv[1:]
print(f'Received {len(args)} arguments.')
for i, arg in enumerate(args):
    print(f'Argument #{i+1}: {arg}')
```
