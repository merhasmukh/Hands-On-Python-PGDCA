## Part 1: Running and diagnosing the broken scripts
Let's inspect the files in `broken_scripts/` directory.


```python
from pathlib import Path
for f in Path('broken_scripts').iterdir():
    print('File:', f.name)
    print('-'*20)
    # Run the file inside subprocess to inspect crash message
    import subprocess
    res = subprocess.run(['python3', str(f)], capture_output=True, text=True)
    print('Stdout:\n', res.stdout)
    if res.stderr:
        print('Stderr:\n', res.stderr)
    print('='*40)
```
