## Part 1: Bulk File Reading with pathlib
Let's list all files in a directory and read their contents inside our script.


```python
from pathlib import Path

data_dir = Path('data')
txt_files = list(data_dir.glob('*.txt'))
for f in txt_files:
    with open(f, 'r', encoding='utf-8') as file:
        content = file.read()
        print(f"File: {f.name} | Characters: {len(content)}")
```

## Hands-On Exercise
**Task**: Write a python script `bulk_process.py` that reads all files matching `*.txt` in the current folder, strips whitespaces, converts to lowercase, and creates a flat summary log output.


```python
%%writefile bulk_process.py
from pathlib import Path

for p in Path('.').glob('*.txt'):
    if p.name == 'temp_emoji.txt': continue
    with open(p, 'r', encoding='utf-8') as f:
        txt = f.read().strip().lower()
    print(f'Processed {p.name}: {len(txt)} chars')
```
