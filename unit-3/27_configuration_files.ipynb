# Practical P27: Configuration Files for Prompts and Parameters
**Learning Outcome**: Read external JSON configuration files to change API parameters dynamically.

## Part 1: Loading config files
Let's write a standard configurations JSON, load it, and use it inside a python dict.


```python
%%writefile config.json
{
    "model": "gpt-4o-mini",
    "temperature": 0.3,
    "max_tokens": 120
}
```

```python
import json

with open('config.json', 'r') as f:
    config = json.load(f)

print('Loaded Model configuration:', config['model'])
print('Temperature threshold:', config['temperature'])
```

## Hands-On Exercise
**Task**: Create a script `run_with_config.py` that reads settings from `config.json` and updates parameters based on optional command-line flags. If `--model` is provided on the command line, overwrite the JSON config model.


```python
%%writefile run_with_config.py
import argparse
import json

parser = argparse.ArgumentParser()
parser.add_argument('--model')
args = parser.parse_args()

with open('config.json', 'r') as f:
    cfg = json.load(f)

if args.model:
    cfg['model'] = args.model
    
print('Current active config model:', cfg['model'])
```
