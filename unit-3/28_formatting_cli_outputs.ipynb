# Practical P28: Formatting CLI Outputs
**Learning Outcome**: Display professional console feedback using tabulate, tqdm progress bars, and structured indicators.

## Part 1: Formatting tabular data
We can print collections of list-of-dicts formatted as a console grid table.


```python
try:
    from tabulate import tabulate
except ImportError:
    # Fallback printer
    def tabulate(data, headers, tablefmt):
        return '\n'.join([str(row) for row in data])
        
results = [
    ['1', 'summarize', 'Success', '0.002$'],
    ['2', 'classify', 'Failed (429)', '0.000$'],
    ['3', 'email', 'Success', '0.004$']
]
headers = ['Job ID', 'Task', 'Status', 'Cost']
print(tabulate(results, headers=headers, tablefmt='grid'))
```

## Part 2: Loading Status indicators and progress bars
Let's simulate processing records with a delay using progress indicators.


```python
import time
try:
    from tqdm import tqdm
    print('Using tqdm progress tracker:')
    for i in tqdm(range(10)):
        time.sleep(0.1)
except ImportError:
    # Simple printable dots fallback
    print('Processing items: ', end='')
    for i in range(10):
        print('.', end='', flush=True)
        time.sleep(0.1)
    print(' Done!')
```
