# Practical P29: CLI Tool: AI Text Summarizer
**Learning Outcome**: Read files, construct prompts dynamically based on `--length` parameters, execute mock AI queries, and support verbose logging operations.

## Part 1: Creating `ai_summarize.py`
Let's write a complete script that implements the requested argparse parameters:


```python
%%writefile ai_summarize.py
import argparse
import logging
import sys

def summarize_text(text, length, model):
    logging.debug(f'summarize_text called with model={model}, length={length}')
    # Simulate summaries depending on length parameter
    if length == 'short':
        return 'AI summaries save developer integration overhead.'
    elif length == 'medium':
        return 'Enterprise AI applications use REST endpoints instead of bespoke ML pipelines, saving deployment overhead.'
    else:
        return 'Modern enterprise platforms leverage cloud LLM completions APIs. This minimizes time-to-market but requires orchestrations to manage API security and errors.'

def main():
    parser = argparse.ArgumentParser(description='AI Summarizer CLI Tool')
    parser.add_argument('--file', required=True, help='Path to target text file')
    parser.add_argument('--length', choices=['short', 'medium', 'long'], default='medium', help='Summary size')
    parser.add_argument('--verbose', action='store_true', help='Enable debug logs')
    parser.add_argument('--model', default='openai', choices=['openai', 'anthropic'], help='AI Model choice')
    
    args = parser.parse_args()
    
    # Configure Logging level based on --verbose
    log_level = logging.DEBUG if args.verbose else logging.INFO
    logging.basicConfig(level=log_level, format='%(asctime)s - %(levelname)s - %(message)s')
    logger = logging.getLogger('Summarizer')
    
    logger.info(f'Reading file {args.file} using model {args.model}')
    try:
        with open(args.file, 'r', encoding='utf-8') as f:
            content = f.read()
    except FileNotFoundError:
        logger.error(f'File not found: {args.file}')
        sys.exit(1)
        
    summary = summarize_text(content, args.length, args.model)
    print('\n--- SUMMARY OUTPUT ---')
    print(summary)
    print('----------------------')

if __name__ == '__main__':
    main()
```
