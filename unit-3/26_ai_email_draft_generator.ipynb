## Part 1: Standard Templates and Replacements
Let's define email formats using f-strings and replace fields dynamically.


```python
email_template = """
Subject: {subject}
To: {recipient}

Dear {recipient},

Thank you for contacting us. Regarding your request: "{request_text}",
here is the automated AI response update:
"{ai_response}"

Best Regards,
Customer Support AI
"""

formatted = email_template.format(
    subject='Billing Query Update',
    recipient='Alice Smith',
    request_text='Refund request for yesterday.',
    ai_response='We have processed your refund of $50 successfully.'
)
print(formatted)
```

## Hands-On Exercise
**Task**: Create a CLI script `email_gen.py` that accepts arguments `--to`, `--subject`, and `--issue`. Load a template, format the email draft and print it.


```python
%%writefile email_gen.py
import argparse

parser = argparse.ArgumentParser()
parser.add_argument('--to', required=True)
parser.add_argument('--subject', required=True)
parser.add_argument('--issue', required=True)

args = parser.parse_args()
template = f"To: {args.to}\nSubject: {args.subject}\n\nHello,\nWe are working on your issue: '{args.issue}'.\nRegards,\nSupport."
print(template)
```
