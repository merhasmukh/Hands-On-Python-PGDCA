## Part 1: Designing an ArgumentParser
Python's standard `argparse` module makes it easy to write user-friendly command-line interfaces. It automatically generates help and usage messages.


```python
import argparse

# Set up parser
parser = argparse.ArgumentParser(description='AI Playground Settings CLI')
# Add positional argument
parser.add_argument('prompt', type=str, help='The text prompt to submit')
# Add optional arguments
parser.add_argument('--temp', type=float, default=0.7, help='Generation temperature')
parser.add_argument('--verbose', action='store_true', help='Enable verbose output')

# Test parsing programmatically (passing custom list of args)
args = parser.parse_args(['Hello AI!', '--temp', '0.2', '--verbose'])
print('Parsed prompt:', args.prompt)
print('Parsed temp:', args.temp)
print('Verbose status:', args.verbose)
```

## Hands-On Exercise
**Task**: Write a CLI script `calc_tokens.py` that takes a text prompt as a positional argument, a `--multiplier` optional integer parameter (default 1), and counts words. Multiply the word count by multiplier and print the output.


```python
%%writefile calc_tokens.py
import argparse

parser = argparse.ArgumentParser(description='Simple Word Token Count Calculator')
parser.add_argument('text', type=str, help='Input text prompt')
parser.add_argument('--multiplier', type=int, default=1, help='Scale factor')

args = parser.parse_args()
word_count = len(args.text.split())
final_count = word_count * args.multiplier
print(f'Original words: {word_count} | Scaled total: {final_count}')
```
