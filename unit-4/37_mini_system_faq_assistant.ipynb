## Part 1: FAQ Assistant Program Structure
We load 20 real FAQs. If user prompt matches a key exactly, load it. Otherwise, fallback to LLM request with context injection.


```python
%%writefile faq_assistant.py
import json
import sys

def find_exact_match(user_query, faq_list):
    query_clean = user_query.strip().lower()
    for faq in faq_list:
        if faq['question'].lower() in query_clean or query_clean in faq['question'].lower():
            return faq['answer']
    return None

def query_ai_fallback(user_query, faq_list):
    # Inject FAQs content into LLM context block
    context = "\n".join([f"Q: {f['question']} -> A: {f['answer']}" for f in faq_list[:5]])
    mock_answer = f"[AI Fallback Answer] Based on FAQ details, the query relates to curriculum structures."
    return mock_answer

def main():
    with open('data/faq_kb.json', 'r') as f:
        faqs = json.load(f)
        
    print('FAQ assistant is online. Type "exit" to stop.')
    print('Ask details about PGDCA duration, curriculum, Python topics, or Streamlit.')
    
    # Pre-loading mock choices for non-interactive test run
    mock_inputs = ['How long is the program?', 'What is Streamlit?', 'exit']
    for q in mock_inputs:
        print(f'\nQuery: {q}')
        if q.lower() == 'exit':
            print('Stopping FAQ service. Goodbye!')
            break
            
        match = find_exact_match(q, faqs)
        if match:
            print(f'Answer (Database Match): {match}')
        else:
            ans = query_ai_fallback(q, faqs)
            print(f'Answer (AI Fallback): {ans}')

if __name__ == '__main__':
    main()
```
