## Part 1: Loading CSV and Analytical Aggregations
We load sales figures, aggregate revenue metrics with pandas, and generate summaries using a prompt mapping.


```python
import pandas as pd
import json

# Load sales CSV
df = pd.read_csv('data/sales_data.csv')
df['units_sold'] = pd.to_numeric(df['units_sold'])
df['unit_price'] = pd.to_numeric(df['unit_price'])
df['revenue'] = df['units_sold'] * df['unit_price']

total_sales = df['revenue'].sum()
avg_units = df['units_sold'].mean()
best_product = df.groupby('product')['units_sold'].sum().idxmax()

stats = {
    'total_sales_usd': int(total_sales),
    'average_units_sold': round(avg_units, 2),
    'best_selling_product': best_product
}
print('Aggregated Stats Summary:\n', json.dumps(stats, indent=2))
```

## Part 2: Combining context into AI prompt report structures
Let's write a mockup report builder.


```python
prompt = f"Write a 1-sentence sales summary for regional performance based on these statistics: {json.dumps(stats)}"
# Mock AI Output summary
ai_executive_summary = f"Our best selling product this month was '{best_product}', contributing to a total sales volume of ${total_sales} USD."

report_text = f"""
==========================================
          MONTHLY SALES REPORT
==========================================
Total Sales: ${total_sales}
Average Units: {avg_units:.1f}
Top Product: {best_product}
------------------------------------------
Executive Narrative summary:
\"{ai_executive_summary}\"
=========================================="""
print(report_text)
```
