# Practical P34: Git & GitHub for AI Projects
**Learning Outcome**: Version control AI projects with Git and publish to GitHub.

## Part 1: Setting up .gitignore
It is crucial to create a `.gitignore` file before running your first `git add` to prevent tracking files like local database weights, packages folder, logs, or secrets.


```python
%%writefile .gitignore
# Environment variables containing secrets
.env

# Python caching and compilation objects
__pycache__/
*.pyc
*.pyo
*.pyd
.ipynb_checkpoints/

# Dataset logs and local database files
*.log
data/raw/
temp_*
```

## Part 2: Git Initialization Commands
In your terminal console, run these sequences to set up tracking:
1. Initialize repo: `git init`
2. Track files: `git add .`
3. Commit changes: `git commit -m "Initial commit containing notebooks and resources"`
4. Link remote hosted server: `git remote add origin <your-github-url>`
5. Push branch: `git push -u origin main`
