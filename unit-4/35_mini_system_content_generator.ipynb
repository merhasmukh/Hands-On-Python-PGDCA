# Practical P35: Mini System 1: AI Content Generator
**Learning Outcome**: Build a configurable AI content generation system.

## Part 1: Configurable Prompt Construction
Let's write a python program `content_generator.py` that accepts topic and options, queries a mock model, formats results, and displays reading statistics.


```python
%%writefile content_generator.py
import argparse
import json
import time

def build_generation_prompt(topic, tone, format_type, max_words):
    return f"Write a {format_type} about '{topic}'. Tone: {tone}. Limit: {max_words} words."

def query_mock_generator(prompt):
    # Mock model response generation
    words = prompt.split()
    result_text = f"[Creative Content] Here is a completed writeup about '{words[4]}'. It is crafted with a highly specialized tone."
    return result_text

def main():
    parser = argparse.ArgumentParser()
    parser.add_argument('--topic', required=True)
    parser.add_argument('--tone', choices=['formal', 'casual', 'creative'], default='casual')
    parser.add_argument('--format', choices=['blog', 'email', 'social'], default='blog')
    parser.add_argument('--words', type=int, default=50)
    
    args = parser.parse_args()
    
    prompt = build_generation_prompt(args.topic, args.tone, args.format, args.words)
    print(f'Sending prompt: "{prompt}"')
    
    content = query_mock_generator(prompt)
    word_count = len(content.split())
    reading_time_sec = int(word_count / 3) + 1 # 180 words per minute estimation
    
    print('\n' + '='*40)
    print('       GENERATED ARTIFACT OUTPUT')
    print('='*40)
    print(content)
    print('='*40)
    print(f'Word Count: {word_count} words | Est. Reading Time: {reading_time_sec}s')
    
    # Export outputs
    with open('data/generated_content.txt', 'w') as f:
        f.write(content)
    
    meta = {'topic': args.topic, 'tone': args.tone, 'format': args.format, 'word_count': word_count}
    with open('data/generated_metadata.json', 'w') as f:
        json.dump(meta, f, indent=2)

if __name__ == '__main__':
    main()
```
