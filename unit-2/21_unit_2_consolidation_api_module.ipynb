## Part 1: Writing the Modular Client Client
Let's write a file named `ai_client.py` directly using the `%%writefile` utility cell.


```python
%%writefile ai_client.py
# ai_client.py - Production-ready AI Client
import os
import time
import logging
from dotenv import load_dotenv

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger('AIClientModule')

class AIClientModule:
    """
    A production-grade module to handle prompt queries, error fallbacks, and cost analytics.
    """
    def __init__(self):
        load_dotenv()
        self.api_key = os.getenv('OPENAI_API_KEY')
        if not self.api_key:
            logger.warning('OPENAI_API_KEY environment variable is not defined!')
            
    def send_prompt(self, prompt, model='gpt-4o-mini'):
        # Check length limit
        if not prompt or len(prompt.strip()) == 0:
            raise ValueError('Prompt cannot be empty.')
            
        logger.info(f'Sending prompt request: "{prompt[:30]}..." to model: {model}')
        time.sleep(0.1) # Mock Network delay
        
        # Mock Response
        response_text = f"AI response to: '{prompt}'"
        total_tokens = len(prompt.split()) + 15
        estimated_cost = (total_tokens / 1000.0) * 0.002
        
        return {
            'status': 'success',
            'model': model,
            'response': response_text,
            'metrics': {
                'tokens': total_tokens,
                'cost_usd': estimated_cost
            }
        }
```
