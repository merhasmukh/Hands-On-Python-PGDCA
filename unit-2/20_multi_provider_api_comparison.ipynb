# Practical P20: Calling Multiple AI APIs & Comparing Responses
**Learning Outcome**: Build a multi-provider AI client and compare results programmatically.

## Part 1: Enhancing the Client for Provider Routing
We will design a unified system where we can change our provider (e.g. `'openai'` vs `'anthropic'`) and the class takes care of matching the correct API payload schemas automatically.


```python
class MultiProviderClient:
    def __init__(self, openai_key, anthropic_key):
        self.openai_key = openai_key
        self.anthropic_key = anthropic_key
        
    def call_provider(self, provider, prompt):
        # Mocking endpoints behaviors side-by-side
        if provider == 'openai':
            # Simulate OpenAI response format
            return {
                'provider': 'openai',
                'latency': 0.15,
                'response': f"[OpenAI GPT-4] I think: '{prompt}'",
                'cost': 0.00035
            }
        elif provider == 'anthropic':
            # Simulate Anthropic response format
            return {
                'provider': 'anthropic',
                'latency': 0.22,
                'response': f"[Anthropic Claude] My analysis shows: '{prompt}'",
                'cost': 0.00050
            }
        else:
            raise ValueError(f'Provider {provider} not supported')
```

## Part 2: Programmatic Execution and Benchmarking
Let's compare responses and latency side by side.


```python
client = MultiProviderClient('sk-mock-op', 'sk-mock-ant')
prompt = 'Define API Integration.'

openai_res = client.call_provider('openai', prompt)
anthropic_res = client.call_provider('anthropic', prompt)

print('--- COMPARISON DASHBOARD ---')
print(f"OpenAI Response:    {openai_res['response']}")
print(f"OpenAI Latency:     {openai_res['latency']}s")
print(f"Anthropic Response: {anthropic_res['response']}")
print(f"Anthropic Latency:  {anthropic_res['latency']}s")
```

## Hands-On Exercise
**Task**: Write a python function `generate_comparison_markdown(prompt)` that calls both providers and returns a clean Markdown table comparing Provider, Response Text, Latency, and Cost.


```python
# TODO: Implement generate_comparison_markdown
def generate_comparison_markdown(prompt):
    client = MultiProviderClient('sk-op', 'sk-ant')
    r_op = client.call_provider('openai', prompt)
    r_an = client.call_provider('anthropic', prompt)
    
    md_table = [
        '| Provider | Response | Latency (s) | Cost ($) |',
        '| :--- | :--- | :--- | :--- |',
        f"| OpenAI | {r_op['response']} | {r_op['latency']} | {r_op['cost']:.5f} |",
        f"| Anthropic | {r_an['response']} | {r_an['latency']} | {r_an['cost']:.5f} |"
    ]
    return '\n'.join(md_table)

from IPython.display import Markdown, display
table = generate_comparison_markdown('Why do developers love Python?')
display(Markdown(table))
```
