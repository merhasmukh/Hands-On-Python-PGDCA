# Practical P12: requests Library: GET & POST
**Learning Outcome**: Make GET and POST requests confidently using the `requests` library.

## Part 1: Installing and Importing `requests` & GET Operations
The `requests` library is the most popular way to send HTTP requests in Python.
Let's fetch public weather data using open-meteo (which does not require API credentials).


```python
import requests

# Fetch coordinates weather for Ahmedabad (Latitude: 23.0225, Longitude: 72.5714)
url = 'https://api.open-meteo.com/v1/forecast'
params = {
    'latitude': 23.0225,
    'longitude': 72.5714,
    'current_weather': True
}

response = requests.get(url, params=params)

print('Status Code:', response.status_code)
print('Is request successful?', response.ok)
print('Content-Type Header:', response.headers.get('Content-Type'))
```

## Part 2: Parsing the GET Response
We can view the response body as plain text (`.text`) or parse it directly as JSON (`.json()`).


```python
# Inspect raw text vs parsed dictionary
data = response.json()
print('Type of parsed response:', type(data))
current = data['current_weather']
print(f"Ahmedabad Temperature: {current['temperature']}°C")
print(f"Windspeed: {current['windspeed']} km/h")
```

## Part 3: POST Requests with JSON Payload
POST requests are used to send data to a server (like sending a prompt to an AI model). We pass headers to specify format and auth details, and a JSON body.


```python
# Simulating a POST call to a mock endpoint (httpbin.org is a free service for testing HTTP requests)
post_url = 'https://httpbin.org/post'
headers = {
    'Content-Type': 'application/json',
    'Authorization': 'Bearer mock-token-12345'
}
payload = {
    'model': 'gpt-4o-mini',
    'prompt': 'Tell me a joke about Python.'
}

post_response = requests.post(post_url, headers=headers, json=payload)
print('POST Response Status:', post_response.status_code)
# Inspect echo values returned by httpbin
echo_data = post_response.json()
print('Sent Headers parsed by server:', echo_data['headers'])
print('Sent JSON payload parsed by server:', echo_data['json'])
```

```python
# TODO: Fetch weather details for Ahmedabad and print a report
url = 'https://api.open-meteo.com/v1/forecast'
params = {
    'latitude': 23.0225,
    'longitude': 72.5714,
    'current_weather': True
}
resp = requests.get(url, params=params)
weather_data = resp.json()['current_weather']

print('='*40)
print('       AHMEDABAD WEATHER DASHBOARD')
print('='*40)
print(f"Time of Reading:  {weather_data['time']}")
print(f"Current Temp:     {weather_data['temperature']}°C")
print(f"Wind Speed:       {weather_data['windspeed']} km/h")
print(f"Wind Direction:   {weather_data['winddirection']}°")
print('='*40)
```
