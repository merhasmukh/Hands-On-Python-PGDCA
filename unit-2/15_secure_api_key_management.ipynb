# Practical P15: Secure API Key Management with dotenv
**Learning Outcome**: Manage API credentials securely using environment variables and block commits of credentials to Git.

## Part 1: Fast-Failing Key Validation at Initialization
When starting an AI application script, check that required API keys exist before running network operations. If keys are missing, crash early with a meaningful error message.


```python
import os
import sys
from dotenv import load_dotenv

def check_config_loaded():
    load_dotenv() # Load variables from local .env file
    
    required_keys = ['OPENAI_API_KEY', 'ANTHROPIC_API_KEY']
    missing = []
    for k in required_keys:
        val = os.getenv(k)
        if not val or val.strip() == '':
            missing.append(k)
            
    if missing:
        raise ValueError(f'Startup Failure: Missing environment keys: {missing}')
    print('All keys are configured correctly.')

try:
    check_config_loaded()
except ValueError as err:
    print('Expected verification error:', err)
```
